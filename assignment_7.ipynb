## Bivariate analysis:

## Bivariate analysis implementation:

### Fixed acidity vs pH:
From the below plot between pH on y axis and fixed acidity on x axis, we may conclude that acidity of wine may vary as decrease in pH range or vice versa .
1. f.acidity and pH are having a correlation coefficient of '-0.43', Hence there relation is inversely related
2. more acidic the wine have less pH value.
3. Wines are having more fixed acidity

```python
sns.regplot(data =df, x='fixed acidity',y='pH',x_jitter=0.2)
```

### Quality vs Alcohol
    From the below plot treds are:
    1. alcohol and quality have correlation coefficient of 0.44
    2. There is more alcohol concentration with higher quality
    3. Average quality wines with quality 5,6 and 7 having alcohol range 10 to 12.
    

```python
sns.regplot(data=df, x='quality',y='alcohol',x_jitter=0.2)
```

### Citric acid vs pH
1. The bivariate plot between these two looks similar as those of fixed acidity and pH.
2. More value of citric acid as les of pH one
3. correlation coefficient between them is -0.16 which is totally in versely


```python
sns.regplot(data = df, x='citric acid',y='pH',color='red')
```

### Density vs Quality
1. correlation coefficient between them is -0.78.
2. more the density less will be the quality.
3. Mediocre quality are densed with range of 4-6 of quality and density within 0.99 to 1.00 

```python
sns.regplot(data=df,x='density',y='quality',y_jitter=0.3,color='green')
```

## Multivariate analysis:

### Density vs alcohol using FacetGrid
1. With the help of FacetGrid , we will be able to draw quality vs density of each qaulity number i.e 3,4..8
2. Through this we can overcome the too overshadow problem as it comes under multivariate analysis. 

```python
grid = sns.FacetGrid(data=df,row='quality')
grid.map(plt.scatter,'density','alcohol')
```

### Quality on behalf of density vs pH
    From the below plots of quality on the behalf of density and ph, the qualtiy somehow increases with less in pH value as density doen't impact anymore all lies in the same range

```python
grid = sns.FacetGrid(data=df,col='quality',col_wrap=2)
grid.map(plt.scatter,'density','pH',alpha=0.1,color='green')
```

### Quality on the basis of alcohol and Chlorides
From the below plot , as the quality of wine increases but the domain facotrs of other doesn't have any more impact on it.
Only the the wines with quality have the higher alcohol level whereas the chloride part remains the same

```python
grid = sns.FacetGrid(data=df,col='quality',col_wrap=3)
grid.map(plt.scatter,'chlorides','alcohol',color='pink',alpha=0.1)
```

### Quality on behalf of Citric acid and pH
From the below plot , as the quality of wine increases upto the mediocre from 5 to 7 with lower pH value as amount of Citric acid increses but decrases with last few qualities

```python
grid = sns.FacetGrid(data=df,col='quality',col_wrap=2)
grid.map(plt.scatter,'citric acid','pH')
```
