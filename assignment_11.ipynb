```
#importing libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

#load the dataset
data = pd.read_csv('https://raw.githubusercontent.com/Shreyas3108/Titanic-EDA-and-Survival-prediction/master/train.csv')



#first five rows of dataset
data.head()

```

NO null values in** Survived**

```
#check for null values
data.isnull().sum()
```

```
#Function for input values to Has_cabin from cAbin 
def func(col):
  if pd.isnull(col):
    return 0
  else:
    return 1
```

```
data['Has_cabin'] = data['Cabin'].apply(func)
```

**Filling null values**

```
#input missing values with fillna()
data.mean()
data.fillna(data.mean())
data.head()
```

Adding column** Categorical_age and Categorical_fare**

```
#categorical_age column on the behalf of age divisions
data['Categorical_age'] = pd.qcut(data['Age'],3,labels = False)
data.head()
```

```
#categorical_fare column using qcut
data['Categorical_fare'] = pd.qcut(data['Fare'],3,labels = [1,2,3])
data.head()
```

Conversion of Variables into Numeric values using ***get_dummies***

```
#transforming variables into numeric values using get_dummies

df=pd.get_dummies(data['Sex'])
data1=pd.concat([data,df],axis=1)
data1.head()
```

```
#Droping columns Cabin,name,passenger,Grid,Ticket,Age,Fare

data2 = data1.drop(columns=['Cabin','Name','PassengerId','Ticket','Fare','SibSp','Age','Parch'])
data.head()
```

```
data2.head()
```

```
from sklearn.linear_model import LogisticRegression
```

```
reg = LogisticRegression()
```

```
#now splitting the dataset to train and test
from sklearn.cross_validation import train_test_split
```

```
labels = data2['Survived']
#declaring variables as features
train1 =data2.drop(['Sex','Survived','Embarked','Categorical_age','Categorical_fare'],axis=1)
```

```
#splitting of dataset as train and test
x_train,x_test,y_train,y_test = train_test_split(train1 , labels , test_size = 0.30,random_state = 2)
```

```

#del data['Sex']
data2.head()
```

```
train1.head

#train1.info()
```

**Training of data**

```

#fitting of training set into model
reg.fit(x_train, y_train)
```

**Accuracy **of Model

```
reg.score(x_test , y_test)
```
