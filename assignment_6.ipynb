Question-1.** Visualisation of any given or provided Data**

```
# First Plot -> "Bar Plot"
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

#dataset used here is weather
df = pd.read_csv("https://raw.githubusercontent.com/Shreyas3108/Titanic-EDA-and-Survival-prediction/master/test.csv")
#df.head
#df.groupby(by = 'Date')
df.head
```

```
#bar plot
#Plot no.1.
c=sns.barplot(y='Fare',x='Sex',data=df)
plt.show(c)
sns.despine()
```

```
#bar plot
#Plot n0.2
sns.barplot(y='Age',x='Embarked',data=df,hue='Sex')
```

```
#Plot no. 3
g = sns.factorplot(x='Parch', y='PassengerId',
                    data=df, kind='bar')
```

```
#Plot no. 4
#Box plot
c = sns.boxplot(data=df)
plt.show(c)
```

```
#plot no. 5
sns.boxplot(x='Sex',y='PassengerId',data=df)
```

```
#plot no .6
sns.boxplot(data=df,x='Parch',y='PassengerId',hue='Sex')
```

```
#plot no. 7
#using linewidth of value 2
sns.boxplot(data=df,x='Parch',y='PassengerId',linewidth='2')
```

```
#plot no. 8
#box plot using swarmplot
sns.boxplot(data=df,x='Sex',y='Age')
sns.swarmplot(data=df,x='Sex',y='Age',color='0')
```

```
#plot no. 9
#using factorplot
sns.factorplot(data=df,x='Age',y='Sex',hue='Parch',kind='box')
```

```
#plot no-10
#Scatterplot
sns.pairplot(data=df,hue='Parch')
```

```
#plot no. 11
#using regplot
sns.regplot(data=df,x='PassengerId',y='Age')
```

```
#plot no. 12
#regplot using jitter
sns.regplot(data=df,x='PassengerId',y='Parch',x_jitter=0.1)
```

```
#plot no. 13
#Using violin plot
sns.violinplot(data=df,x='Age',y='Sex',hue='Parch')
```

```
#plot no. -14
#violin plot part -2
sns.violinplot(data=df,x='Parch',y='Age',split=True,hue='Sex')
```
