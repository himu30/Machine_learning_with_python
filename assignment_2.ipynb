```
#question-1  calculating mean of array
import numpy as np
a=np.random.random((10,1))
print(a.shape)
print(np.mean(a))
```

```
#question - 2 calculating variance and deviation
b=np.random.random((20,1))
print(b.shape)
print("Standard deviation :",np.std(b))
print("Variance :",np.var(b))
```

```
#question-3
#matrix multiplication
a=np.random.rand(10,20)
b=np.random.rand(20,25)
c=np.dot(a,b)
print(c)
print("Shape of new matrix is : ",c.shape)
print("Sum of elements of new array is:",np.sum(c))
```
